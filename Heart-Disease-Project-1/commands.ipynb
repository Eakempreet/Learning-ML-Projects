## 1) The Most Useful Commands

```powershell
# Check conda
conda --version
conda info

# Create a project environment in the current folder
conda create --prefix ./.conda python=3.11 -y

# Activate the environment
conda activate ./.conda

# Install common data science packages
conda install numpy pandas scikit-learn matplotlib seaborn -y

# Install from conda-forge
conda install -c conda-forge jupyterlab -y

# See installed packages
conda list

# Leave the environment
conda deactivate
```

<span style="color:#60a5fa;">blue</span> = information, 
<span style="color:#4ade80;">green</span> = install, 
<span style="color:#f87171;">red</span> = remove/delete.

## 2) Name-Based vs Path-Based Environments

| Method | Example | Best For | Tradeoff |
|---|---|---|---|
| Name-based | `conda create -n myenv python=3.11` | Shared personal environments | Not tied to one project |
| Path-based | `conda create --prefix ./.conda python=3.11` | Project-specific work | Slightly longer activate command |

### Recommendation

Use **path-based** environments for project work.

Why this is better:
- The environment lives inside the project.
- It is easier to remember which project it belongs to.
- It is easier to share or move the project later.
- It keeps your work cleaner and more organized.

## 4) Install, Update, and Search

```powershell
# Search for a package
conda search xgboost

# Install a specific version
conda install pandas=2.2.2 -y

# Update one package
conda update pandas -y

# Update everything in the active environment
conda update --all -y
```

## 5) Removing Packages and Deleting Environments

<div style="padding:10px 12px;border-radius:10px;background:#2b1d14;border:1px solid #fb923c;color:#fde68a;">
This is the part people often confuse.
</div>

Think of it like this:
- <span style="color:#4ade80;"><strong>remove</strong></span> means take one item out of the room.
- <span style="color:#f87171;"><strong>delete</strong></span> means remove the whole room.

```powershell
# Remove a single package from the active environment
conda remove seaborn -y

# Remove multiple packages
conda remove seaborn xgboost -y

# Delete an environment by name
conda env remove -n myenv

# Delete a project environment by path
conda env remove --prefix ./.conda

# Clean up downloaded package caches
conda clean --all -y
```

If you want to completely rebuild a broken environment, deleting it and creating it again is often the cleanest fix.

## 6) GitHub / Git Commands

If you are using GitHub for your project, these are the main commands you will use.

```powershell
# Check status
git status

# See recent history
git log --oneline

# Start tracking a folder if needed
git init

# Add files to staging
git add .

# Save a commit
git commit -m "Initial commit"

# Connect to GitHub repository
git remote add origin https://github.com/username/repo-name.git

# Send changes to GitHub
git push -u origin main

# Get latest changes from GitHub
git pull origin main

# See your branches
git branch

# Create and switch to a new branch
git checkout -b feature-branch
```

### Useful GitHub workflow

1. Make your changes.
2. Run `git status` to see what changed.
3. Use `git add .` to stage files.
4. Use `git commit -m "message"` to save a checkpoint.
5. Use `git push` to send it to GitHub.

## 7) Channels

```powershell
# Show channels
conda config --show channels

# Add conda-forge
conda config --add channels conda-forge
conda config --set channel_priority strict
```

Simple rule:
- Prefer <span style="color:#60a5fa;"><strong>conda-forge</strong></span> for data science packages when possible.
- Try to keep most packages from the same ecosystem in one environment.

## 8) Export and Recreate

```powershell
# Export the full environment
conda env export > environment.yml

# Export only the packages you explicitly installed
conda env export --from-history > environment.history.yml

# Recreate from a file
conda env create -f environment.yml

# Recreate by path
conda env create --prefix ./.conda -f environment.yml
```

---

## 9) Pip vs Conda

Use this order inside a Conda environment:
1. Install with <span style="color:#4ade80;"><strong>conda</strong></span> first.
2. Use <span style="color:#c084fc;"><strong>pip</strong></span> only if the package is not available in conda channels.

```powershell
conda activate ./.conda
conda install -c conda-forge numpy pandas -y
pip install some-package-not-on-conda
```

Rule:
- Avoid mixing many random `pip` and `conda` installs.
- If things get messy, recreate the environment from a clean file.

## 10) Troubleshooting

```powershell
# If conda is acting slow or confused
conda clean --all -y

# Check the active environment path
echo $env:CONDA_PREFIX

# Check which Python is being used
python -c "import sys; print(sys.executable)"
```

If activation fails in PowerShell:
```powershell
conda init powershell
# Restart the terminal after this
```
