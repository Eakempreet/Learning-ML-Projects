# Conda / Anaconda Terminal Cheat Sheet

A quick, project-focused guide you can actually use daily.

---

## 1) The 80/20 Commands (Use These Most)

```powershell
# Check conda
conda --version
conda info

# Create project env at a path (recommended)
conda create --prefix ./.conda python=3.11 -y

# Activate project env by path
conda activate ./.conda

# Install packages
conda install numpy pandas scikit-learn -y

# Install from conda-forge channel
conda install -c conda-forge jupyterlab -y

# List installed packages
conda list

# Deactivate env
conda deactivate
```

---

## 2) Create Env: Name vs Path (Which One to Prefer)

| Method | Command Style | Best For | Tradeoff |
|---|---|---|---|
| Name-based | `conda create -n myenv python=3.11` | Shared personal envs | Not tied to one project |
| Path-based | `conda create --prefix ./.conda python=3.11` | Project-specific work | Slightly longer activate command |

## Recommendation

Use **path-based** envs for project work.

Why it is better for you:
- Environment lives inside the project.
- No confusion about which env belongs to which project.
- Easier to move/share project setup.
- Matches your preference for project-specific isolation.

---

## 3) Project Workflow (Recommended)

```powershell
# From project root
conda create --prefix ./.conda python=3.11 -y
conda activate ./.conda
conda install -c conda-forge numpy pandas matplotlib scikit-learn jupyterlab -y
```

Optional: keep the environment folder hidden-ish with a dot name like `.conda` or `.env-conda`.

---

## 4) Package Management Essentials

```powershell
# Search a package
conda search xgboost

# Install exact version
conda install pandas=2.2.2 -y

# Update one package
conda update pandas -y

# Update all packages in active env
conda update --all -y

# Remove a package
conda remove seaborn -y
```

---

## 5) Channels (Very Important)

```powershell
# Show current channels
conda config --show channels

# Add conda-forge at top priority
conda config --add channels conda-forge
conda config --set channel_priority strict
```

Practical rule:
- Prefer one ecosystem per env when possible.
- `conda-forge` is often the most complete for data science.

---

## 6) Export, Recreate, Share

```powershell
# Export full environment (can be OS-specific)
conda env export > environment.yml

# Export only explicitly requested packages (cleaner)
conda env export --from-history > environment.history.yml

# Recreate env from file by name
conda env create -f environment.yml

# Recreate env by path
conda env create --prefix ./.conda -f environment.yml
```

---

## 7) Useful Advanced Commands

```powershell
# Clone an env
conda create --name myenv-clone --clone myenv -y

# List all envs
conda env list

# Remove env by name
conda env remove -n myenv

# Remove env by path
conda env remove --prefix ./.conda

# Clean unused caches (safe cleanup)
conda clean --all -y
```

---

## 8) When to Use pip vs conda

Use this order inside a Conda env:
1. Install what you can with `conda` first.
2. Use `pip` only for packages not available on conda channels.

```powershell
conda activate ./.conda
conda install -c conda-forge numpy pandas -y
pip install some-package-not-on-conda
```

Rule:
- Avoid mixing many `pip` and `conda` installs randomly.
- If env gets messy, recreate from a clean `environment.yml`.

---

## 9) Troubleshooting Quick Fixes

```powershell
# If solver is slow/confused
conda clean --all -y

# Check active env path
echo $env:CONDA_PREFIX

# Verify Python executable being used
python -c "import sys; print(sys.executable)"
```

If activation fails in PowerShell:
```powershell
conda init powershell
# Restart terminal after this
```

## 10) Copy-Paste Templates

### A) New ML project (path-based, recommended)

```powershell
conda create --prefix ./.conda python=3.11 -y
conda activate ./.conda
conda install -c conda-forge numpy pandas matplotlib scikit-learn jupyterlab notebook -y
```

### B) Fast recreate in same project later

```powershell
conda env create --prefix ./.conda -f environment.history.yml
conda activate ./.conda
```

### C) Remove and rebuild project env

```powershell
conda env remove --prefix ./.conda
conda create --prefix ./.conda python=3.11 -y
conda activate ./.conda
```

## My Suggested Standard for Your Projects

Use this every time:
1. `conda create --prefix ./.conda python=3.11 -y`
2. `conda activate ./.conda`
3. Install with `conda` first, `pip` only when needed.
4. Save `environment.history.yml` for clean reproducibility.

This keeps each project clean, isolated, and easy to recover.
